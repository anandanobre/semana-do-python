# WEB SCRAPING COM SELENIUM
Automação para web scraping no goggle chrome ultilizando selenium

## Extraindo as cotações

```python
# Importanto bibliotecas necessárias
from selenium import webdriver # Automatiza a a web(Roda em segudo plano)
from selenium.webdriver.common.keys import Keys # Permite escrever textos no navegador

# Abre o navegador
nav = webdriver.Chrome()
```

```python
# Pesquisa a cotação do dólar e pega o valor
nav.get("https://www.google.com.br/")

# Encontra o elemente barra de pesquisa pelo xPath dela
# send_keys envia o texto para o elemento e da enter
bar = nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
bar.send_keys("cotação dólar")
bar.send_keys(Keys.ENTER)

# Pegando valor da busca
val = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')
c_dolar = val.get_attribute('data-value')
print(c_dolar)
```

```python
# Pesquisa a cotação do euro e pega o valor
nav.get("https://www.google.com.br/")

# Encontra o elemente barra de pesquisa pelo xPath dela
# send_keys envia o texto para o elemento e da enter
bar = nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
bar.send_keys("cotação euro")
bar.send_keys(Keys.ENTER)

# Pegando valor da busca
val = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')
c_euro = val.get_attribute('data-value')
print(c_euro)
```

```python
# Pesquisa a cotação do ouro e pega o valor
nav.get("https://www.melhorcambio.com/ouro-hoje")
val = nav.find_element_by_xpath('//*[@id="comercial"]')
c_ouro = val.get_attribute('value')
c_ouro = c_ouro.replace(",", ".") # Troca a virgula por um ponto
print(c_ouro)
```

## Atualizando a base de dados

```python
import pandas as pd
df = pd.read_excel("Produtos.xlsx")
display(df)
```

### Atualizando cotações

```python
# Atualizando valores
# Localiza as linhas e colunas a serem editadas df.loc[linhas, colunas] = valor
df.loc[df["Moeda"] == "Dólar", "Cotação"] = float(c_dolar)

df.loc[df["Moeda"] == "Euro", "Cotação"] = float(c_euro)

df.loc[df["Moeda"] == "Ouro", "Cotação"] = float(c_ouro)

display(df)
```

### Atualizando preço base

```python
df["Preço Base Reais"] = df["Cotação"] * df["Preço Base Original"]

display(df)
```

### Atualizando preço final

```python
df["Preço Final"] = df["Preço Base Reais"] * df["Ajuste"]

display(df)
```

### Formatando número da tabela


```python
df["Cotação"] = df["Cotação"].map("{:.2f}".format)

df["Preço Base Reais"] = df["Preço Base Reais"].map("{:.2f}".format)

df["Preço Final"] = df["Preço Final"].map("{:.2f}".format)

display(df)
```

## Exportando base de dados atualizada

```python
# Cria nova tabela com os valores atualizados e removendo os indices
df.to_excel("ProdutosAtualizados.xlsx", index = False)
```
